```python
# !pip install faker holidays requests -q

import pandas as pd
import numpy as np
import random
import string
!pip install Faker
from faker import Faker
import holidays
from datetime import datetime, timedelta
import os

fake = Faker('th_TH')   # Thai locale for realistic names/addresses
np.random.seed(42)
random.seed(42)

OUTPUT_DIR = "mock_data"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ─── Shared config ────────────────────────────
N_PRODUCTS    = 80
N_CUSTOMERS   = 500
N_STORES      = 10
N_WAREHOUSES  = 3
N_PROMOTIONS  = 30
START_DATE    = datetime(2023, 1, 1)
END_DATE      = datetime(2024, 12, 31)
DATE_RANGE    = pd.date_range(START_DATE, END_DATE, freq='D')

PRODUCT_CATEGORIES = [
    'beverages', 'snacks', 'dairy', 'frozen_food',
    'personal_care', 'household', 'bakery', 'condiments'
]
STORE_TYPES    = ['urban_large', 'suburban_medium', 'rural_small', 'convenience']
CUSTOMER_SEGS  = ['high_value', 'regular', 'occasional', 'new']
WAREHOUSE_TYPES = ['central', 'regional_north', 'regional_south']

def rand_id(prefix, n, width=4):
    return [f"{prefix}-{str(i).zfill(width)}" for i in range(1, n+1)]

product_ids   = rand_id('SKU',   N_PRODUCTS)
customer_ids  = rand_id('CUST',  N_CUSTOMERS)
store_ids     = rand_id('STORE', N_STORES)
warehouse_ids = rand_id('WH',    N_WAREHOUSES)
promo_ids     = rand_id('PROMO', N_PROMOTIONS)

print("✅ Config ready")
```

```python
"""
WHY: product_taxonomies enables grouped/hierarchical forecasting.
     When a single SKU has sparse sales, we fall back to category-level.
     base_price is used to compute revenue and gross margin.
"""

category_map = {pid: random.choice(PRODUCT_CATEGORIES) for pid in product_ids}

product_master = pd.DataFrame({
    'product_id':         product_ids,
    'product_name':       [f"{cat.replace('_',' ').title()} Item {i+1}"
                           for i, (pid, cat) in enumerate(category_map.items())],
    'product_taxonomies': list(category_map.values()),
    'base_price':         np.round(
                              np.random.choice([9.0,15.0,19.0,25.0,35.0,49.0,
                                                69.0,89.0,99.0,149.0],
                                              size=N_PRODUCTS), 2),
    'unit_of_measure':    np.random.choice(['piece','bottle','pack','kg','litre'],
                                           size=N_PRODUCTS),
    'shelf_life_days':    np.random.choice([7,14,30,60,180,365], size=N_PRODUCTS),
})

product_master.to_csv(f"{OUTPUT_DIR}/product_master.csv", index=False)
print(f"✅ product_master: {len(product_master)} rows")
print(product_master.head(3).to_string())
```

```python
"""
WHY: store_taxonomies lets us train separate sub-models per store type.
     Urban large stores have very different demand patterns from rural small.
     region is used to join weather data (different regions = different temps).
"""

store_master = pd.DataFrame({
    'store_id':          store_ids,
    'store_name':        [f"Branch {s}" for s in store_ids],
    'store_taxonomies':  np.random.choice(STORE_TYPES, size=N_STORES),
    'region':            np.random.choice(['Bangkok','Central','North','South'],
                                          size=N_STORES),
    'opening_year':      np.random.choice(range(2015, 2023), size=N_STORES),
    'size_sqm':          np.random.choice([80,150,300,500,800], size=N_STORES),
})

store_master.to_csv(f"{OUTPUT_DIR}/store_master.csv", index=False)
print(f"✅ store_master: {len(store_master)} rows")
```

```python
"""
WHY: customer_taxonomies seeds the RFM segmentation with prior labels.
     acquisition_channel tells us HOW to deliver promotions (LINE vs walk-in).
     is_loyalty_member is used as a feature weight in the uplift model.
"""

customer_master = pd.DataFrame({
    'customer_id':          customer_ids,
    'customer_taxonomies':  np.random.choice(CUSTOMER_SEGS,
                                             p=[0.15,0.45,0.25,0.15],
                                             size=N_CUSTOMERS),
    'acquisition_channel':  np.random.choice(['LINE_OA','walk_in','online'],
                                             p=[0.50,0.35,0.15],
                                             size=N_CUSTOMERS),
    'is_loyalty_member':    np.random.choice([True, False],
                                             p=[0.40, 0.60],
                                             size=N_CUSTOMERS),
    'registration_date':    pd.to_datetime(
                                np.random.choice(pd.date_range('2021-01-01',
                                                               '2024-01-01'),
                                                 size=N_CUSTOMERS)),
    'province':             np.random.choice(
                                ['Bangkok','Chiang Mai','Chonburi','Khon Kaen'],
                                size=N_CUSTOMERS),
})

customer_master.to_csv(f"{OUTPUT_DIR}/customer_master.csv", index=False)
print(f"✅ customer_master: {len(customer_master)} rows")
```

```python
warehouse_master = pd.DataFrame({
    'warehouse_id':          warehouse_ids,
    'warehouse_taxonomies':  WAREHOUSE_TYPES,
    'location':              ['Bangkok','Chiang Mai','Surat Thani'],
    'capacity_units':        [50000, 20000, 15000],
    'cold_storage':          [True, False, True],
})

warehouse_master.to_csv(f"{OUTPUT_DIR}/warehouse_master.csv", index=False)
print(f"✅ warehouse_master: {len(warehouse_master)} rows")
```

```python
"""
WHY: discount + start/end date → promo_active feature for S1 forecasting.
     budget_thb + cost_per_redemption → enables true ROI calculation for S2.
     Without cost columns, we can only measure sales lift (vanity metric).
"""

promo_records = []
for pid in promo_ids:
    start = START_DATE + timedelta(days=random.randint(0, 600))
    duration = random.choice([3, 7, 14, 21, 30])
    end = start + timedelta(days=duration)
    n_products = random.randint(1, 5)
    for prod in random.sample(product_ids, n_products):
        promo_records.append({
            'promotion_id':        pid,
            'product_id':          prod,
            'discount':            round(random.choice([0.05,0.10,0.15,0.20,0.25,0.30]), 2),
            'start_date':          start.date(),
            'end_date':            end.date(),
            'budget_thb':          random.choice([10000,25000,50000,100000]),
            'cost_per_redemption': round(random.uniform(5, 30), 2),
            'channel_cost_thb':    random.choice([0, 2000, 5000, 10000]),
            'target_segment':      random.choice(CUSTOMER_SEGS),
        })

promotion_master = pd.DataFrame(promo_records)
promotion_master.to_csv(f"{OUTPUT_DIR}/promotion_master.csv", index=False)
print(f"✅ promotion_master: {len(promotion_master)} rows")
```

```python
"""
WHY: po_date → arrival_date gives us supplier lead time (critical for reorder point).
     expire_date is the CORE signal for Solution 3 (Markdown Engine).
     po_price_per_unit gives cost basis for gross margin calculation.
"""

po_records = []
po_ids = rand_id('PO', 800)

for po_id in po_ids:
    po_date = START_DATE + timedelta(days=random.randint(0, 600))
    lead_time = random.randint(2, 14)           # days from order to arrival
    arrival = po_date + timedelta(days=lead_time)
    mfg_date = po_date - timedelta(days=random.randint(1, 30))
    product = random.choice(product_ids)
    shelf_life = int(product_master.loc[
        product_master['product_id']==product, 'shelf_life_days'].values[0])
    expire = arrival + timedelta(days=shelf_life)
    qty = random.randint(50, 500)
    base_price = float(product_master.loc[
        product_master['product_id']==product, 'base_price'].values[0])
    cost_per_unit = round(base_price * random.uniform(0.40, 0.65), 2)

    po_records.append({
        'po_id':              po_id,
        'warehouse_id':       random.choice(warehouse_ids),
        'product_id':         product,
        'po_date':            po_date.date(),
        'arrival_date':       arrival.date(),
        'manufacturing_date': mfg_date.date(),
        'expire_date':        expire.date(),
        'qty':                qty,
        'unit':               'piece',
        'po_price_per_unit':  cost_per_unit,
        'po_price_total':     round(cost_per_unit * qty, 2),
        'lead_time_days':     lead_time,    # derived column — saves compute later
    })

purchasing_order = pd.DataFrame(po_records)
purchasing_order.to_csv(f"{OUTPUT_DIR}/purchasing_order.csv", index=False)
print(f"✅ purchasing_order: {len(purchasing_order)} rows")
print(f"   expire_date range: {purchasing_order['expire_date'].min()} → {purchasing_order['expire_date'].max()}")
```

```python
"""
WHY: Without this, we cannot compute current stock per store.
     current_stock = sum(stock_movement.qty) - sum(sales.qty)
     This is the denominator in days_of_stock = current_stock / avg_daily_demand.
"""

stock_records = []
for _, po in purchasing_order.iterrows():
    n_transfers = random.randint(1, 3)
    remaining = po['qty']
    transfer_date = pd.to_datetime(po['arrival_date']) + timedelta(days=random.randint(1,3))

    for t in range(n_transfers):
        if remaining <= 0:
            break
        if t == n_transfers - 1:
            qty_t = remaining
        else:
            qty_t = random.randint(10, max(11, remaining - 10))
        remaining -= qty_t
        store = random.choice(store_ids)
        receive = transfer_date + timedelta(days=random.randint(0, 2))

        stock_records.append({
            'po_id':          po['po_id'],
            'warehouse_id':   po['warehouse_id'],
            'store_id':       store,
            'transfer_date':  transfer_date.date(),
            'receive_date':   receive.date(),
            'qty':            qty_t,
        })
        transfer_date += timedelta(days=random.randint(0, 3))

stock_movement = pd.DataFrame(stock_records)
stock_movement.to_csv(f"{OUTPUT_DIR}/stock_movement.csv", index=False)
print(f"✅ stock_movement: {len(stock_movement)} rows")
```

```python
"""
WHY: The primary demand signal for ALL three solutions.
     We simulate realistic Thai retail patterns:
       - Payday spikes (end of month)
       - Weekend bumps
       - Songkran / New Year demand multipliers
       - Promotion-driven demand lift
"""

th_holidays = holidays.Thailand(years=[2023, 2024])

def get_demand_multiplier(date, product_cat, promo_active):
    mult = 1.0
    if date.weekday() >= 5:                      # weekend
        mult *= 1.20
    if date.day >= 28 or date.day <= 2:          # payday window
        mult *= 1.35
    if date in th_holidays:
        name = th_holidays[date].lower()
        if 'songkran' in name:
            mult *= (2.5 if product_cat in ['beverages','snacks'] else 1.4)
        else:
            mult *= 1.3
    if promo_active:
        mult *= 1.45
    return mult

# Build a promo lookup: (product_id, date) → is_promo_active
promo_lookup = {}
for _, row in promotion_master.drop_duplicates(['promotion_id','product_id']).iterrows():
    dates = pd.date_range(row['start_date'], row['end_date'])
    for d in dates:
        promo_lookup[(row['product_id'], d.date())] = row['discount']

# Base daily demand per product (units sold per day on average)
base_demand = {pid: random.randint(2, 25) for pid in product_ids}

sales_records = []
sale_idx = 1

for date in DATE_RANGE:
    for store in store_ids:
        # Sample 5–15 products sold this day at this store
        daily_products = random.sample(product_ids, k=random.randint(5, 15))
        for product in daily_products:
            cat = category_map[product]
            promo = (product, date.date()) in promo_lookup
            discount = promo_lookup.get((product, date.date()), 0)
            mult = get_demand_multiplier(date, cat, promo)
            qty = max(1, int(np.random.poisson(base_demand[product] * mult)))
            base_price = float(product_master.loc[
                product_master['product_id']==product, 'base_price'].values[0])
            sell_price = round(base_price * (1 - discount), 2)
            customer = random.choice(customer_ids)
            po_ref = random.choice(po_ids)
            promo_ref = None
            if promo:
                matches = promotion_master[
                    (promotion_master['product_id']==product) &
                    (promotion_master['start_date'] <= date.date()) &
                    (promotion_master['end_date']   >= date.date())
                ]['promotion_id'].values
                if len(matches):
                    promo_ref = matches[0]

            sales_records.append({
                'sale_id':      f"SALE-{str(sale_idx).zfill(6)}",
                'datetime':     date + timedelta(hours=random.randint(8,21),
                                                 minutes=random.randint(0,59)),
                'product_id':   product,
                'store_id':     store,
                'customer_id':  customer,
                'qty':          qty,
                'price':        sell_price,
                'revenue':      round(sell_price * qty, 2),
                'promotion_id': promo_ref,
                'po_id':        po_ref,
            })
            sale_idx += 1

sales_transaction = pd.DataFrame(sales_records)

if 'sales_transaction' in locals() or 'sales_transaction' in globals():
    # แปลงคอลัมน์ datetime ให้เป็นประเภท datetime object ของ pandas ก่อน
    sales_transaction['datetime'] = pd.to_datetime(sales_transaction['datetime'])
    sales_transaction['date'] = sales_transaction['datetime'].dt.date

    # 3.1 จำลองตารางวันหยุดนักขัตฤกษ์ของไทย (ตัวอย่างปี 2026)
    th_holidays = [
    # ---- ปี 2023 ----
    '2023-01-01', '2023-01-02', # วันขึ้นปีใหม่ / ชดเชย
    '2023-03-06',               # วันมาฆบูชา
    '2023-04-06',               # วันจักรี
    '2023-04-13', '2023-04-14', '2023-04-15', '2023-04-17', # วันสงกรานต์ / ชดเชย
    '2023-05-01',               # วันแรงงาน
    '2023-05-04',               # วันฉัตรมงคล
    '2023-06-03', '2023-06-05', # วันวิสาขบูชา / วันเฉลิมพระชนมพรรษาพระราชินี / ชดเชย
    '2023-07-28',               # วันเฉลิมพระชนมพรรษารัชกาลที่ 10
    '2023-08-01', '2023-08-02', # วันอาสาฬหบูชา / วันเข้าพรรษา
    '2023-08-12', '2023-08-14', # วันแม่แห่งชาติ / ชดเชย
    '2023-10-13',               # วันคล้ายวันสวรรคตรัชกาลที่ 9
    '2023-10-23',               # วันปิยมหาราช
    '2023-12-05',               # วันพ่อแห่งชาติ
    '2023-12-10', '2023-12-11', # วันรัฐธรรมนูญ / ชดเชย
    '2023-12-31',               # วันสิ้นปี

    # ---- ปี 2024 ----
    '2024-01-01',               # วันขึ้นปีใหม่
    '2024-02-24', '2024-02-26', # วันมาฆบูชา / ชดเชย
    '2024-04-06', '2024-04-08', # วันจักรี / ชดเชย
    '2024-04-13', '2024-04-14', '2024-04-15', '2024-04-16', # วันสงกรานต์ / ชดเชย
    '2024-05-01',               # วันแรงงาน
    '2024-05-04', '2024-05-06', # วันฉัตรมงคล / ชดเชย
    '2024-05-22',               # วันวิสาขบูชา
    '2024-06-03',               # วันเฉลิมพระชนมพรรษาพระราชินี
    '2024-07-20', '2024-07-22', # วันอาสาฬหบูชา / ชดเชย
    '2024-07-28', '2024-07-29', # วันเฉลิมพระชนมพรรษารัชกาลที่ 10 / ชดเชย
    '2024-08-12',               # วันแม่แห่งชาติ
    '2024-10-13', '2024-10-14', # วันคล้ายวันสวรรคตรัชกาลที่ 9 / ชดเชย
    '2024-10-23',               # วันปิยมหาราช
    '2024-12-05',               # วันพ่อแห่งชาติ
    '2024-12-10',               # วันรัฐธรรมนูญ
    '2024-12-31'                # วันสิ้นปี
]
    th_holidays = [pd.to_datetime(d).date() for d in th_holidays]

    # สร้างฟีเจอร์ วันหยุดนักขัตฤกษ์ (1 = ใช่, 0 = ไม่ใช่)
    sales_transaction['is_holiday'] = sales_transaction['date'].isin(th_holidays).astype(int)

    # สร้างฟีเจอร์ วันเสาร์-อาทิตย์ (5=เสาร์, 6=อาทิตย์)
    sales_transaction['is_weekend'] = sales_transaction['datetime'].dt.dayofweek.isin([5, 6]).astype(int)

    # 3.2 จำลองสภาพอากาศ (is_raining: 1 = ฝนตกหนัก, 0 = ตกเบาหรือแดดออก)
    # อิงตามความเป็นจริง: เดือน มิ.ย. - ต.ค. (เดือน 6-10) โอกาสฝนตกจะสูงกว่าเดือนอื่น
    sales_transaction['month'] = sales_transaction['datetime'].dt.month

    def simulate_weather(month):
        if month in [6, 7, 8, 9, 10]:  # หน้าฝน
            return np.random.choice([0, 1], p=[0.60, 0.40]) # โอกาสฝนตก 40%
        else:
            return np.random.choice([0, 1], p=[0.90, 0.10]) # โอกาสฝนตก 10%

    sales_transaction['is_raining'] = sales_transaction['month'].apply(simulate_weather)

sales_transaction.to_csv(f"{OUTPUT_DIR}/sales_transaction.csv", index=False)
print(f"✅ sales_transaction: {len(sales_transaction):,} rows")
print(f"   Date range: {sales_transaction['datetime'].min()} → {sales_transaction['datetime'].max()}")
print(f"   Total revenue: {sales_transaction['revenue'].sum():,.0f} THB")
```

```python
import holidays # Ensure holidays is imported

"""
WHY: Without this, the model treats Songkran as unexplained anomalies.
     Adding holiday flags + countdown features reduces MAPE by ~15-30%
     on seasonal SKUs (beverages, snacks, gifts).
     Source: holidays Python library (official Thai calendar).
"""

th_holidays = holidays.Thailand(years=[2023, 2024]) # Re-initialize th_holidays as the holidays object

holiday_rows = []
for date in DATE_RANGE:
    is_holiday = date.date() in th_holidays
    name = th_holidays.get(date.date(), None)
    # Days to next holiday
    future = [d for d in th_holidays.keys() if d >= date.date()]
    days_to_next = (min(future) - date.date()).days if future else 999
    # Weight: major holidays get higher multiplier
    weight = 1.0
    if is_holiday:
        if name and 'songkran' in name.lower():
            weight = 2.5
        elif name and ('new year' in name.lower() or 'christmas' in name.lower()):
            weight = 2.0
        else:
            weight = 1.5

    holiday_rows.append({
        'date':                date.date(),
        'is_public_holiday':   is_holiday,
        'holiday_name':        name,
        'days_to_next_holiday': days_to_next,
        'holiday_demand_weight': weight,
        'is_weekend':          date.weekday() >= 5,
        'is_payday_window':    date.day >= 28 or date.day <= 2,
    })

holiday_calendar = pd.DataFrame(holiday_rows)
holiday_calendar.to_csv(f"{OUTPUT_DIR}/holiday_calendar.csv", index=False)
print(f"✅ holiday_calendar: {len(holiday_calendar)} rows")
print(f"   Holidays in range: {holiday_calendar['is_public_holiday'].sum()}")
```

```python
"""
WHY: Thai FMCG demand is temperature-sensitive.
     Beverages spike 20-40% on hot days (≥35°C).
     Cold medicine, umbrellas spike on rainy days.
     Source: Can fetch real data from OpenMeteo API (free, no key).
     Here we simulate realistic Bangkok seasonal patterns.

     To fetch real data instead:
       import requests
       url = "https://api.open-meteo.com/v1/forecast"
       params = {"latitude":13.75,"longitude":100.5,
                 "daily":"temperature_2m_max,precipitation_sum",
                 "start_date":"2023-01-01","end_date":"2024-12-31",
                 "timezone":"Asia/Bangkok"}
       r = requests.get(url, params=params).json()
"""

def simulate_bangkok_temp(date):
    """Realistic Bangkok temperature by month (hot season Apr-May, cooler Nov-Feb)"""
    month = date.month
    base_temp = {1:28,2:30,3:33,4:36,5:35,6:33,
                 7:32,8:32,9:31,10:31,11:30,12:28}[month]
    return round(base_temp + np.random.normal(0, 2), 1)

def simulate_rainfall(date):
    """Rainy season Jun-Oct for Bangkok"""
    month = date.month
    rainy_months = [5,6,7,8,9,10]
    if month in rainy_months:
        return round(max(0, np.random.exponential(15)), 1)
    return round(max(0, np.random.exponential(2)), 1)

weather_rows = []
for date in DATE_RANGE:
    temp = simulate_bangkok_temp(date)
    rain = simulate_rainfall(date)
    weather_rows.append({
        'date':             date.date(),
        'region':           'Bangkok',
        'temp_max_c':       temp,
        'precipitation_mm': rain,
        'is_hot_day':       temp >= 35,
        'is_rainy_day':     rain >= 10,
    })

weather_data = pd.DataFrame(weather_rows)
weather_data.to_csv(f"{OUTPUT_DIR}/weather_data.csv", index=False)
print(f"✅ weather_data: {len(weather_data)} rows")
print(f"   Hot days (≥35°C): {weather_data['is_hot_day'].sum()}")
print(f"   Rainy days (≥10mm): {weather_data['is_rainy_day'].sum()}")
```

```python
"""
WHY: Not all products respond the same to discounts.
     Beverages are elastic (big demand jump from small discount).
     Branded goods are inelastic (10% off barely moves volume).
     Without this, the markdown model uses the same % for everything = wrong.
     Elasticity can be derived from Sales Transaction or estimated from literature.
"""

elasticity_data = pd.DataFrame([
    {'product_taxonomies': 'beverages',     'price_elasticity': -2.1, 'recommended_max_discount': 0.30, 'source': 'derived'},
    {'product_taxonomies': 'snacks',        'price_elasticity': -1.8, 'recommended_max_discount': 0.25, 'source': 'derived'},
    {'product_taxonomies': 'dairy',         'price_elasticity': -1.5, 'recommended_max_discount': 0.30, 'source': 'literature'},
    {'product_taxonomies': 'frozen_food',   'price_elasticity': -1.3, 'recommended_max_discount': 0.35, 'source': 'derived'},
    {'product_taxonomies': 'personal_care', 'price_elasticity': -0.8, 'recommended_max_discount': 0.20, 'source': 'literature'},
    {'product_taxonomies': 'household',     'price_elasticity': -0.7, 'recommended_max_discount': 0.15, 'source': 'literature'},
    {'product_taxonomies': 'bakery',        'price_elasticity': -1.9, 'recommended_max_discount': 0.40, 'source': 'derived'},
    {'product_taxonomies': 'condiments',    'price_elasticity': -0.9, 'recommended_max_discount': 0.20, 'source': 'literature'},
])
# Interpretation: elasticity of -2.1 means a 1% price decrease → 2.1% demand increase
elasticity_data.to_csv(f"{OUTPUT_DIR}/price_elasticity.csv", index=False)
print(f"✅ price_elasticity: {len(elasticity_data)} rows")
```

```python
"""
WHY: Quantifies the "before" state so ROI of Solution 3 is tangible.
     When client sees "you lost 180,000 THB last year from expired stock",
     they immediately understand the value of the markdown engine.
     Mock: assume 2-5% of PO qty expires per category.
"""

waste_records = []
for _, po in purchasing_order.iterrows():
    product = po['product_id']
    cat = category_map[product]
    # Short shelf life categories waste more
    waste_rate_map = {
        'bakery':0.08,'dairy':0.06,'frozen_food':0.05,
        'beverages':0.03,'snacks':0.02,'condiments':0.02,
        'personal_care':0.01,'household':0.01
    }
    waste_rate = waste_rate_map.get(cat, 0.03) * random.uniform(0.5, 1.5)
    expired_qty = int(po['qty'] * waste_rate)
    if expired_qty > 0:
        loss = round(expired_qty * po['po_price_per_unit'], 2)
        waste_records.append({
            'po_id':           po['po_id'],
            'product_id':      product,
            'store_id':        random.choice(store_ids),
            'expire_date':     po['expire_date'],
            'expired_qty':     expired_qty,
            'cost_per_unit':   po['po_price_per_unit'],
            'loss_value_thb':  loss,
            'reason':          random.choice(['no_markdown_applied',
                                              'late_markdown',
                                              'demand_lower_than_expected']),
        })

waste_log = pd.DataFrame(waste_records)
waste_log.to_csv(f"{OUTPUT_DIR}/waste_log.csv", index=False)
total_waste = waste_log['loss_value_thb'].sum()
print(f"✅ waste_log: {len(waste_log)} rows")
print(f"   Total waste value: {total_waste:,.0f} THB  ← show this to the client!")
```

```python
import pandas as pd
import numpy as np

# ตั้งค่า Seed เพื่อให้รันกี่ครั้งข้อมูลสุ่มก็ยังได้ค่าเดิม (ง่ายต่อการทำซ้ำใน Colab)
np.random.seed(42)

# ==========================================
# 1. เพิ่มข้อมูลต้นทุน (COGS) ใน Product Master
# ==========================================
# ทางธุรกิจ: สินค้าแต่ละกลุ่มจะมี Gross Profit Margin ไม่เท่ากัน
# สมมติฐาน: ต้นทุน (COGS) มักจะอยู่ที่ 50% ถึง 70% ของราคาขาย (Price)
if 'product_master' in locals() or 'product_master' in globals():
    # สร้างอัตราส่วนต้นทุนแบบสุ่มระหว่าง 0.5 - 0.7 ให้สินค้าแต่ละชิ้น
    cogs_ratio = np.random.uniform(0.5, 0.7, size=len(product_master))
    product_master['cogs'] = (product_master['base_price'] * cogs_ratio).round(2)

    product_master.to_csv(f"{OUTPUT_DIR}/product_master.csv", index=False)
    print(" [Success] Added 'cogs' to Product Master")


# ==========================================
# 2. เพิ่มข้อมูล Customer Demographics ใน Customer Master
# ==========================================
# ทางธุรกิจ: ช่วยให้ทำ Customer Segmentation (RFM + Demographics) ได้ลึกขึ้น
if 'customer_master' in locals() or 'customer_master' in globals():
    # จำลองสัดส่วนเพศ (ชาย 45%, หญิง 50%, ไม่ระบุ 5%)
    genders = ['M', 'F', 'Unknown']
    gender_probs = [0.45, 0.50, 0.05]

    # จำลองพื้นที่สาขาหลักของลูกค้า (เน้นกรุงเทพและหัวเมืองใหญ่)
    locations = ['Bangkok', 'Vicinity', 'Chonburi', 'Chiang Mai', 'Khon Kaen', 'Phuket']
    location_probs = [0.50, 0.15, 0.10, 0.10, 0.08, 0.07]

    customer_master['age'] = np.random.randint(18, 65, size=len(customer_master))
    customer_master['gender'] = np.random.choice(genders, size=len(customer_master), p=gender_probs)
    customer_master['location'] = np.random.choice(locations, size=len(customer_master), p=location_probs)

    customer_master.to_csv(f"{OUTPUT_DIR}/customer_master.csv", index=False)
    print(" [Success] Added 'age', 'gender', 'location' to Customer Master")


# ==========================================
# 3. เพิ่มข้อมูล External Factors ใน Sales Transaction
# ==========================================
# ทางธุรกิจ: วันหยุดและสภาพอากาศมีผลอย่างมากต่อ Demand Forecasting (ยอดขายจะพุ่งช่วงวันหยุด/ฝนตกคนเข้าห้าง)
if 'sales_transaction' in locals() or 'sales_transaction' in globals():
    # แปลงคอลัมน์ datetime ให้เป็นประเภท datetime object ของ pandas ก่อน
    sales_transaction['datetime'] = pd.to_datetime(sales_transaction['datetime'])
    sales_transaction['date'] = sales_transaction['datetime'].dt.date

    # 3.1 จำลองตารางวันหยุดนักขัตฤกษ์ของไทย (ตัวอย่างปี 2026)
    th_holidays = [
    # ---- ปี 2023 ----
    '2023-01-01', '2023-01-02', # วันขึ้นปีใหม่ / ชดเชย
    '2023-03-06',               # วันมาฆบูชา
    '2023-04-06',               # วันจักรี
    '2023-04-13', '2023-04-14', '2023-04-15', '2023-04-17', # วันสงกรานต์ / ชดเชย
    '2023-05-01',               # วันแรงงาน
    '2023-05-04',               # วันฉัตรมงคล
    '2023-06-03', '2023-06-05', # วันวิสาขบูชา / วันเฉลิมพระชนมพรรษาพระราชินี / ชดเชย
    '2023-07-28',               # วันเฉลิมพระชนมพรรษารัชกาลที่ 10
    '2023-08-01', '2023-08-02', # วันอาสาฬหบูชา / วันเข้าพรรษา
    '2023-08-12', '2023-08-14', # วันแม่แห่งชาติ / ชดเชย
    '2023-10-13',               # วันคล้ายวันสวรรคตรัชกาลที่ 9
    '2023-10-23',               # วันปิยมหาราช
    '2023-12-05',               # วันพ่อแห่งชาติ
    '2023-12-10', '2023-12-11', # วันรัฐธรรมนูญ / ชดเชย
    '2023-12-31',               # วันสิ้นปี

    # ---- ปี 2024 ----
    '2024-01-01',               # วันขึ้นปีใหม่
    '2024-02-24', '2024-02-26', # วันมาฆบูชา / ชดเชย
    '2024-04-06', '2024-04-08', # วันจักรี / ชดเชย
    '2024-04-13', '2024-04-14', '2024-04-15', '2024-04-16', # วันสงกรานต์ / ชดเชย
    '2024-05-01',               # วันแรงงาน
    '2024-05-04', '2024-05-06', # วันฉัตรมงคล / ชดเชย
    '2024-05-22',               # วันวิสาขบูชา
    '2024-06-03',               # วันเฉลิมพระชนมพรรษาพระราชินี
    '2024-07-20', '2024-07-22', # วันอาสาฬหบูชา / ชดเชย
    '2024-07-28', '2024-07-29', # วันเฉลิมพระชนมพรรษารัชกาลที่ 10 / ชดเชย
    '2024-08-12',               # วันแม่แห่งชาติ
    '2024-10-13', '2024-10-14', # วันคล้ายวันสวรรคตรัชกาลที่ 9 / ชดเชย
    '2024-10-23',               # วันปิยมหาราช
    '2024-12-05',               # วันพ่อแห่งชาติ
    '2024-12-10',               # วันรัฐธรรมนูญ
    '2024-12-31'                # วันสิ้นปี
]
    th_holidays = [pd.to_datetime(d).date() for d in th_holidays]

    # สร้างฟีเจอร์ วันหยุดนักขัตฤกษ์ (1 = ใช่, 0 = ไม่ใช่)
    sales_transaction['is_holiday'] = sales_transaction['date'].isin(th_holidays).astype(int)

    # สร้างฟีเจอร์ วันเสาร์-อาทิตย์ (5=เสาร์, 6=อาทิตย์)
    sales_transaction['is_weekend'] = sales_transaction['datetime'].dt.dayofweek.isin([5, 6]).astype(int)

    # 3.2 จำลองสภาพอากาศ (is_raining: 1 = ฝนตกหนัก, 0 = ตกเบาหรือแดดออก)
    # อิงตามความเป็นจริง: เดือน มิ.ย. - ต.ค. (เดือน 6-10) โอกาสฝนตกจะสูงกว่าเดือนอื่น
    sales_transaction['month'] = sales_transaction['datetime'].dt.month

    def simulate_weather(month):
        if month in [6, 7, 8, 9, 10]:  # หน้าฝน
            return np.random.choice([0, 1], p=[0.60, 0.40]) # โอกาสฝนตก 40%
        else:
            return np.random.choice([0, 1], p=[0.90, 0.10]) # โอกาสฝนตก 10%

    sales_transaction['is_raining'] = sales_transaction['month'].apply(simulate_weather)

    sales_transaction.to_csv(f"{OUTPUT_DIR}/sales_transaction.csv", index=False)
    print(" [Success] Added 'is_holiday', 'is_weekend', 'is_raining' to Sales Transaction")
```

```python
"""
Verify joins work before you start modelling.
A broken join is the #1 silent killer of DS projects.
"""

print("Running join sanity checks...\n")

# Check 1: Sales ← Product
merged = sales_transaction.merge(product_master[['product_id','product_taxonomies']],
                                  on='product_id', how='left')
nulls = merged['product_taxonomies'].isna().sum()
print(f"Sales ← Product join: {nulls} nulls (should be 0)")

# Check 2: PO ← Product
merged2 = purchasing_order.merge(product_master[['product_id','shelf_life_days']],
                                  on='product_id', how='left')
nulls2 = merged2['shelf_life_days'].isna().sum()
print(f"PO ← Product join:    {nulls2} nulls (should be 0)")

# Check 3: Expiry risk scan
today_sim = pd.to_datetime('2024-10-01')
purchasing_order['expire_date'] = pd.to_datetime(purchasing_order['expire_date'])
purchasing_order['days_to_expiry'] = (purchasing_order['expire_date'] - today_sim).dt.days
at_risk = purchasing_order[purchasing_order['days_to_expiry'].between(0, 30)]
print(f"\nExpiry risk scan (as of 2024-10-01):")
print(f"  Products expiring within 30 days: {len(at_risk)} POs")
print(f"  Total at-risk qty: {at_risk['qty'].sum():,} units")
print(f"  At-risk value: {(at_risk['qty'] * at_risk['po_price_per_unit']).sum():,.0f} THB")

print("\n✅ All checks passed — proceed to EDA notebook!")
```
